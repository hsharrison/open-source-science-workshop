## 1. Calculate relative phase

The formula for discrete relative phase is
$$
\phi_j = 2 \pi \frac{ t_{\textrm{max L }j} - t_{\textrm{max R } j}}{ t_{\textrm{max L }j+1} - t_{\textrm{max L }j} },
$$
where $t_{\textrm{max K }i}$ is the time of the $i$th maximum extension of limb $K$, and $\phi_i$ is the relative phase for cycle $i$.

First we need to find the maximum extensions from the 3D data.
Since the data look pretty regular, it would probably be okay to just use the $x$ data.
But I am going to try using principal component analysis to collapse the 3D data into 1D data.

```python
from sklearn.decomposition import PCA

left_1d = PCA(n_components=1).fit_transform(left)
right_1d = PCA(n_components=1).fit_transform(right)

plt.plot(t[:3000], left_1d[:3000],
         t[:3000], right_1d[:3000])
```

```python
def collapse(data):
    return np.squeeze(PCA(n_components=1).fit_transform(data))

left_1d = collapse(left)
right_1d = collapse(right)
```

The next step is to find the peaks in the series.
Our first try is to use the function `scipy.signal.argrelmax`.

```python
from scipy.signal import argrelmax

left_peaks = argrelmax(left_1d[:3000])[0]
plt.plot(t[:3000], left_1d[:3000],
         left_peaks, left_1d[left_peaks], 'ko')
```

Unfortunately there is some noise so we pick up a few extra points.
There is another function `find_peaks_cwt`, also in `scipy`'s `signal` module, that uses smoothing to find only sharp peaks in noisy data.
The first argument is the data, the second argument is the range of timescales that you want it to look for peaks in.

I was having trouble finding the right range of timescales so I decided to learn how to use the IPython `interact` feature to explore the options.

```python
from scipy.signal import find_peaks_cwt
from IPython.html.widgets import interact


def find_and_plot_peaks(low=1, high=10):
    left_peaks = find_peaks_cwt(left_1d, np.arange(low, high))
    right_peaks = find_peaks_cwt(right_1d, np.arange(low, high))
    
    plt.subplot(2, 1, 1)
    plt.plot(t, left_1d,
             left_peaks, left_1d[left_peaks], 'k.')
    plt.subplot(2, 1, 2)
    plt.plot(t, right_1d,
             right_peaks, right_1d[right_peaks], 'k.')
    

interact(find_and_plot_peaks, low=(1, 100), high=(1, 100))
```

That still didn't work so I have a new idea: use the original `argrelmax` method, but eliminate all "peaks" with value greater than 0.

```python
def find_peaks(vector, cutoff=0):
    peak_idxs = argrelmax(vector)[0]
    return peak_idxs[vector[peak_idxs] > cutoff]
```

```python
left_peaks = find_peaks(left_1d)
right_peaks = find_peaks(right_1d)

plt.subplot(2, 1, 1)
plt.plot(t, left_1d,
         left_peaks, left_1d[left_peaks], '.')
plt.subplot(2, 1, 2)
plt.plot(t, right_1d,
         right_peaks, right_1d[right_peaks], '.')
```

Excellent, we're ready to continue.
Next we finally apply the formula

$$
\phi_j = 2 \pi \frac{ t_{\textrm{max L }j} - t_{\textrm{max R } j}}{ t_{\textrm{max L }j+1} - t_{\textrm{max L }j} }.
$$

```python
def discrete_relative_phase(left, right):
    left_peaks = find_peaks(collapse(left), 1.75)
    right_peaks = find_peaks(collapse(right), 1.75)
    
    # Make sure both vectors are the same length.
    min_length = min([left_peaks.shape[0], right_peaks.shape[0]])
    left_peaks = left_peaks[:min_length]
    right_peaks = right_peaks[:min_length]
    
    return 2 * np.pi * (left_peaks - right_peaks)[:-1] / np.diff(left_peaks)
```

```python
def wrap(vector):
    return ((vector + np.pi) % (2*np.pi)) - np.pi

def discrete_relative_phase(left, right):
    left_peaks = find_peaks(collapse(left))
    right_peaks = find_peaks(collapse(right))
    
    # Make sure both vectors are the same length.
    min_length = min([left_peaks.shape[0], right_peaks.shape[0]])
    left_peaks = left_peaks[:min_length]
    right_peaks = right_peaks[:min_length]
    
    return wrap(2 * np.pi * (left_peaks - right_peaks)[:-1] / np.diff(left_peaks))
```

```python
def circular_histogram(data, n_bins=20, ax=None):
    bin_edges = np.linspace(-np.pi, np.pi, n_bins, endpoint=False)
    bar_heights, _ = np.histogram(data, bins=bin_edges) 
    
    if ax is None:
        ax = plt.axes(polar=True)
    ax.bar(bin_edges[:-1], bar_heights, width=2*np.pi/n_bins, bottom=2)
```

```python
from scipy.stats import gaussian_kde

def circular_density(data, n_points=50, ax=None):  
    points_to_evaluate = np.linspace(-np.pi, np.pi, n_points + 1)
    kernel = gaussian_kde(data)  # sns has ways to do a density plot more easily but I wanted to
                                # demonstrate the "manual" way to do a kernel density estimation
    
    if ax is None:
        ax = plt.axes(polar=True)
    ax.plot(points_to_evaluate, kernel.evaluate(points_to_evaluate))
```

```python
def near_zero_density(data, n_points=20, lim=np.pi/4, ax=None):
    points_to_evaluate = np.linspace(-lim, lim, n_points)
    density = gaussian_kde(data).evaluate(points_to_evaluate)
    
    if ax is None:
        ax = plt.axes()
    ax.plot(np.rad2deg(points_to_evaluate), density)
```

# 3. Cluster phase and amplitude

Another way of dealing with circular statistics is to use cluster phase $\theta$ and cluster amplitude $r$, associated with the mean and the variance of the phase relations, respectively.

$\theta$ and $r$ can be calculate with some complex arithmetic, as follows:
\begin{align}
c &= \textrm{E}[e^{iX}] \\
r &= |c| \\
\theta &= \textrm{angle }(c),
\end{align}
where $X$ is the phase timeseries, $\textrm{E}$ is the expectation operator, and the function $\textrm{angle}$ is the angle of a complex number within the complex plane.

An alternative method is
\begin{align}
a &= \textrm{E}[\cos x] \\
b &= \textrm{E}[\sin x] \\
r &= \sqrt{a^2 + b^3} \\
\tan \theta &= \frac{b}{a}.
\end{align}

```python
def cluster_analysis(data):
    c = np.mean(np.exp(1j * data))  # In Python, the imaginary i is made with 1j.
    return np.absolute(c), np.angle(c)
cluster_analysis(rel_phase)
```

```python
def cluster_analysis(data):
    vec = [np.mean(np.sin(data)), np.mean(np.cos(data))]
    return np.linalg.norm(vec), np.arctan2(*vec)    
cluster_analysis(rel_phase)
```

I'll interpret here to demonstrate `nameduple`.
For functions that return more than one value, this is a good way to help keep track of which is which.
It doesn't change anything--it still acts like a tuple.
But it will be labeled.
And you can use dot notation to access items as if they were attributes.

Another way to think of a `nameduple` is as a minimal `Class`.

```python
from collections import namedtuple

ClusterResult = namedtuple('ClusterResult', ('r', 'theta'))
```

```python
def cluster_analysis(data):
    c = np.mean(np.exp(1j * data))
    r = np.absolute(c)
    theta = np.angle(c) % 2 * np.pi   
    return ClusterResult(r, theta)
cluster_analysis(discrete_relative_phase(*load_data('data01.txt')))
```

# 4. Simulation of Langevin equation

Next, we simulate the HKB model
$$
\dot{\phi} = - \Delta \omega - a \sin \phi - 2b \sin 2 \phi + \sqrt{Q \zeta_t},
$$
where $\phi$ is relative phase, $\Delta \omega$ is a detuning parameter, $b/a$ is a frequency parameter, and the last term is a noise term (fluctuating force).

For simplicity we will use
$$
\dot{\phi} = h(\phi) + \sqrt{2Q}\Gamma(t),
$$
where $h(\phi)$ is the deterministc "force"
$$
h(\phi) = \frac{\textrm{d }}{\textrm{d}t} V(\phi) = - \alpha \sin \phi,
$$
where $V$ is the potential function.

We will test the prediction of the model by setting the model parameters from the empirical data, with $\alpha = \theta$ and $Q = r$.

Note: technically to simulate a stochastic DE, you should use a stochastic DE solver.
I have found a couple options but none of them are properly packaged for easy use so I haven't had a chance to test them.
In any case rest assured that Till didn't use a dedicated SDE solver in his class (even though they are available in Matlab).

We will just use a "homemade" Euler method solver--that is, at every timestep we calculate $\dot{\phi}$, and then create a new value of $\phi$ by adding $\dot{\phi}\textrm{ d}t$ to the old value of $\phi$.

```python
from toolz import curry

def solve_langevin_eqn(h_func, x0, t, q):
    x = [x0]
    for dt in np.diff(t):
        deterministic = dt * h_func(x[-1])
        fluctuation = np.sqrt(2*q*dt) * np.random.randn()
        x.append(x[-1] + deterministic + fluctuation)
    return np.array(x)

@curry
def hkb(a, x):
    return -a * np.sin(x)
```

```python
t = np.arange(0, 10, 0.0001)
phi_sim = solve_langevin_eqn(hkb(1), 3*np.pi/4, t, 0.1)
plt.plot(t, np.rad2deg(wrap(phi_sim)))
```

```python
VarianceComparison = namedtuple('VarianceComparison', ('analytical', 'numerical'))

def compare_sim_to_data(relative_phase):
    q, a = cluster_analysis(relative_phase)
    t = np.arange(0, 100, 0.01)
    phi_sim = wrap(solve_langevin_eqn(hkb(a), 0, t, q))
    return VarianceComparison(q/a, np.var(phi_sim))
```

```python
for filename in glob('data*.txt'):
    print(filename)
    print('-'*len(filename))
    phase = discrete_relative_phase(*load_data(filename))
    result = compare_sim_to_data(phase)
    print('q = {:.2f}, a = {:.2f}'.format(*cluster_analysis(phase)))
    print('analytical variance: {:.2f}'.format(result.analytical))
    print('numercal variance  : {:.2f}'.format(result.numerical))
    print()
```

```python
def entropy(data, **kwargs):
    heights, _ = np.histogram(data, density=True, **kwargs)
    heights = heights[np.nonzero(heights)]
    
    return - (heights * np.log(heights)).sum(axis=0)
```

# 6. Potential

```python
from scipy import signal
from functools import partial

def hkb_v(b_a, delta_omega, phi):
    return delta_omega * phi - np.cos(phi) - b_a * np.cos(2 * phi)

def plot_hkb_potential(b_a=1, delta_omega=0):
    v_fcn = partial(hkb_v, b_a, delta_omega)
    phi = np.arange(-np.pi -.2, np.pi + .2, np.pi/100)
    
    v = v_fcn(phi)
    plt.plot(phi, v)
    plt.xlim([-np.pi, np.pi])
    plt.ylim([-2.5, 2])
    
    attractors = signal.argrelmin(v)
    plt.plot(phi[attractors], v[attractors], 'ko')
    
    repellers = signal.argrelmax(v)
    plt.plot(phi[repellers], v[repellers], 'ro')
```

```python
interact(plot_hkb_potential, b_a=(0., 2), delta_omega=(-2., 2.))
```
