```python
means = data.reset_index().groupby(
    ['angle', 'perception', 'participant']).mean()
means['response'] = np.arcsin(np.sqrt(means['can step']))
means.head()
```

```python
r.globalenv['means'] = com.convert_to_r_dataframe(
    means.reset_index(), strings_as_factors=True)
print(r.globalenv['means'])
```

```python
r.r("""
means$angle <- factor(means$angle)
means$participant <- factor(means$participant)
""")
```

```python
for variable in ['response', 'rt', 'confidence']:
    print('\n\n{}\n-------'.format(variable))
    print(r.r('summary(aov({} ~ angle*perception + Error(participant/(angle*perception)), data=means))'.format(variable)))
```
