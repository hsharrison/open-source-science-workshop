```python
import pandas as pd

data = pd.read_table('data.csv', sep=',').set_index(['participant', 'block', 'trial'])
data
```

```python
data.groupby(['angle', 'perception']).mean()
```

```python
means = data.groupby(['angle', 'perception']).mean()
%matplotlib inline
```

```python
means.loc[(slice(None), 'haptic'), 'rt'].plot(kind='bar')
```

```python
data.groupby(['angle', 'perception']).std()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
lm = ols('rt ~ C(angle) *C(perception)', data=data.reset_index()).fit()
```

```python
sm.stats.anova_lm(lm)
```
