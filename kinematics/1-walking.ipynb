```python
def difference(frame, columns=None, in_place=False):
    """Differentiate multiple columns from a data frame.
    Differences are appended with prefix 'd'.
    
    Parameters
    ----------
    frame : DataFrame
        DataFrame to operate on.
    colums : list of str, optional
        Columns to difference.
        If not given, all columns will be differenced.
    in_place : bool, optional
        If True, add columns in-place.
        
    """
    if not in_place:
        frame = frame.copy()
    
    if columns is None:
        columns = frame.columns
        
    for column in columns:
        frame['d' + column] = frame[column].diff()
    
    return frame
```

```python
ankle = difference(ankle)
ankle.head()
```

```python
ankle['displacement'] = np.linalg.norm(ankle.ix[:,('dx', 'dy', 'dz')], axis=1)
```

```python
ankle.head()
```

```python
ankle['speed'] = ankle['displacement'] / ankle['dt']
ankle.head()
```

```python
def process_3d_point(frame, in_place=False):
    """
    Given a DataFrame of a 3D point
    with columns ('x', 'y', 'z', 't'),
    calculate and add columns
    ('dx', 'dy', 'dz', 'dt', 'displacement', 'speed', 'accel').
    
    Parameters
    ----------
    frame : DataFrame
        Data to process.
    in_place : bool, optional
        If True, operate in-place.
    
    """
    if not in_place:
        frame = frame.copy()
        
    difference(frame, ('t', 'x', 'y', 'z'), in_place=True)
    frame['displacement'] = np.linalg.norm(frame.ix[:, ('dx', 'dy', 'dz')], axis=1)
    frame['speed'] = frame['displacement'] / frame['dt']
    frame['accel'] = frame['speed'].diff() / frame['dt']
    
    frame['accel'] = np.roll(frame['accel'], -1)
    
    return frame
```

```python
ankle = df.ix[:, ('Time', 'right_ankle X', 'right_ankle Y', 'right_ankle Z')]
ankle.columns = ['t', 'x', 'y', 'z']
ankle = process_3d_point(ankle)
ankle.head()
```

```python
knee = df.ix[:, ('Time', 'right_knee X', 'right_knee Y', 'right_knee Z')]
knee.columns = ['t', 'x', 'y', 'z']
knee = process_3d_point(knee)
knee.head()
```

```python
leg = knee.ix[:, 'x':'z'] - ankle.ix[:, 'x':'z']
leg[['t', 'dt']] = ankle[['t', 'dt']]
leg.head()
```

```python
leg['angle'] = np.arctan2(leg['z'], np.linalg.norm(leg.ix[:, 'x':'y'], axis=1))
leg.head()
```

```python
np.rad2deg(leg['angle'])
```

```python
two_d_angle = np.arctan2(leg['z'], leg['y'])
np.rad2deg(two_d_angle)
```

```python
leg['length'] = np.linalg.norm(leg.ix[:, 'x':'z'], axis=1)
leg['d_angle_dt'] = (leg['angle'].diff() / leg['length'] )/ leg['dt']
```

```python
%matplotlib inline

p = ankle.plot(x='t', y='displacement')
p
```

```python
leg['dd_angle_ddt'] = np.roll(leg['d_angle_dt'].diff() / leg['dt'], -1)
leg.head()
```

```python
leg.set_index('t').plot(y='dd_angle_ddt')
```
